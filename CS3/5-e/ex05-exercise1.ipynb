```python
import pandas as pd
```

```python
data = "data/user_data.csv"
df = pd.read_csv(data)
```

```python
print(df.shape) 
print(df.info())
display(df.head())
```

```python
df[df.duplicated(keep=False)]
```

```python
df = df.drop_duplicates().reset_index(drop=True)
df
```

```python
df.isnull().sum()
```

```python
df['occupation'] = df['occupation'].fillna('unknown')
df.head()
```

```python
df = df.dropna().reset_index(drop=True)
df
```

```python
df['age'] = df['age'].astype(int)
df
```

```python
df['birth_year'] = 2025 - df['age']
df.head()
```

```python
df = df.drop(columns=['zip_code'])
df.head()
```

```python
df['occupation'].value_counts()
```

```python
df['occupation'] = df['occupation'].replace({
    'eductor': 'educator',
    'Writer': 'writer',
    'write': 'writer'
})
df['occupation'].value_counts()
```

```python
df[(df['age'] < 10) | (df['age'] >= 70)]
```

```python
df[(df['occupation'] == 'artist') & (df['gender'] == 'M')]['age'].mean()
```

```python
df[df['occupation'] != 'none'].reset_index(drop=True)
df
```

```python
df.sort_values(by='age').reset_index(drop=True)
```

```python
print(df['age'].idxmin())
print(df['age'].idxmax())
```

```python
df.to_csv('new_user_data.csv', index=False, encoding='utf-8-sig')
```
