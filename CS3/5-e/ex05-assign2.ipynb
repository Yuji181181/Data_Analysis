```python
import pandas as pd
```

```python
data = "data/customers.csv"
df = pd.read_csv(data)
```

```python
print(df.shape) 
print(df.info())
display(df.head())
```

# １番

```python
df[df.duplicated()]
```

# ２番

```python
df2 = df.drop_duplicates().reset_index(drop=True)
df2.shape[0]
```

# ３番

```python
df2.isnull().sum()
```

# ４番

```python
df2['store'] = df2['store'].fillna('Akabane')
df2.head(12)
```

# ５番

```python
df3 = df2.dropna().reset_index(drop=True)
df3.shape[0]
```

# ６番

```python
df3['store'] = df3['store'].replace({'hakusan': 'Hakusan', 'Hakusn': 'Hakusan'})
df3.head(16)
```

# ７番

```python
df3.to_csv("customers_cleaned.csv", index=False)
```

# 課題３

```python
df3['senior'] = df3['age'].map(lambda x: 1 if x >= 60 else 0)
```
