# plt.barの使い方

`plt.bar`は、Matplotlibライブラリで棒グラフを描画することができる  

## 基本的な使い方

```python
    import matplotlib.pyplot as plt

    # データの用意(一次元)
    x = ['A', 'B', 'C', 'D']
    y = [10, 15, 7, 12]

    # 棒グラフの描画
    plt.bar(x, y)

    # タイトルやラベルの設定
    plt.title('タイトル')
    plt.xlabel('カテゴリ')
    plt.ylabel('値')

    # グラフの表示
    plt.show()
```

## 主な引数

- `x`：棒グラフのx軸（カテゴリや位置）の値
- `height`または`y`：各棒の高さ（値）
- `width`：棒の幅（デフォルトは0.8）
- `color`：棒の色
- `label`：凡例用のラベル

## オプション例

```python
plt.bar(x, y, width=0.5, color='skyblue', label='サンプルデータ')
plt.legend()
```

## 横向きの棒グラフ

横向きの棒グラフを描きたい場合は、`plt.barh`を用いる  

```python
plt.barh(x, y)
plt.show()
```



````python
import matplotlib.pyplot as plt
x_list = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
y_list = [3, 7, 5, 8, 6, 4, 9]
plt.bar(x_list, y_list, color='green', width=0.5)
plt.xlabel('Days of the Week')
plt.ylabel('Number of Tasks')
plt.title('Tasks Completed Each Day')
plt.xticks(rotation=45)
plt.show()
````
